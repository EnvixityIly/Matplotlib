```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
students_names=["sanjay","Rahul","karen","Wasim","Mahesh dalleh","Ajay","Sartaj","PankaJ"]
students_marks=[35,50,20,45,25,40,25,40]
```

```python
marks_perc = []
for x in students_marks:
    res = (x/50)*100
    marks_perc.append(res)
```

```python
def line_chart_of_student_and_marks():
    plt.plot(students_names,students_marks, marker = "*",mec= "b",ms=15,color="purple",linestyle="dotted",linewidth=5)
    plt.title("Students mark graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Marks")
    plt.show()

line_chart_of_student_and_marks()

    
```

```python
def bar_perc():
    plt.bar(students_names,marks_perc, color="black")
    plt.title("Students mark percentage graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students percentage")
    plt.show()

bar_perc()
```
